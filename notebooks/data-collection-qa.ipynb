```python
import base64
import json
from io import BytesIO

from PIL import Image
```

```python
data = json.load(open('../data-collection/2024-09-14-19:34:04-annotated-d4.json'))
```

```python
i = 1
```

```python
obs, state, instruction, action = data['history'][i]
if len(obs['error_prefix']) > 0 and not obs['clean_axtree_txt'].startswith(
    obs['error_prefix']
):
    axtree = f"{obs['error_prefix']}\n" + obs['clean_axtree_txt']
else:
    axtree = obs['clean_axtree_txt']
image_data = base64.b64decode(obs['screenshot'])
screenshot = Image.open(BytesIO(image_data))
new_size = (1280, 720)
resized_screenshot = screenshot.resize(new_size)
print('Goal:', data['goal'])
display(resized_screenshot)
print('Observation:\n' + axtree)
print('State:\n' + state)
print('Instruction:\n' + instruction)
print('Action:\n' + action)
```

```python
i += 1
```
